# Pre Processing

```python
#checking for null values
df.isna().sum()
```

```python
#using mode to fill categorical terms

df['gender'] = df["gender"].fillna(df['gender'].mode()[0])
df['weight'] = df["weight"].fillna(df['weight'].mode()[0])
df['strong_foot'] = df["strong_foot"].fillna(df['strong_foot'].mode()[0])
df['coaching'] = df["coaching"].fillna(df['coaching'].mode()[0])
```

```python
#using mean to fill numerical terms

df['ball_controlling_skills'] = df['ball_controlling_skills'].fillna(df['ball_controlling_skills'].mean())
df['jumping_skills'] = df['jumping_skills'].fillna(df['jumping_skills'].mean())
df['penalties_conversion_rate'] = df['penalties_conversion_rate'].fillna(df['penalties_conversion_rate'].mean())
df['mental_strength'] = df['mental_strength'].fillna(df['mental_strength'].mean())
df['shot_accuracy'] = df['shot_accuracy'].fillna(df['shot_accuracy'].mean())
df['behaviour_rating'] = df['behaviour_rating'].fillna(df['behaviour_rating'].mean())
df['matches_played'] = df['matches_played'].fillna(df['matches_played'].mean())
df['fitness_rating'] = df['fitness_rating'].fillna(df['fitness_rating'].mean())
df['years_of_experience'] = df['years_of_experience'].fillna(df['years_of_experience'].mean())
```

```python
from sklearn.preprocessing import LabelEncoder
cols = ["player_id","name","weight","gender","strong_foot","coaching","country","height","dedication_level",]
le = LabelEncoder()
for col in cols:
    df[col] = le.fit_transform(df[col])
```

```python
#matches_played has large no of outliers so removing them

Q1 = np.percentile(df['matches_played'],25,method = 'midpoint')
Q2 = np.percentile(df['matches_played'],50,method = 'midpoint')
Q3 = np.percentile(df['matches_played'],75,method = 'midpoint')

print("q1 is",Q1)
print("q2 is",Q2)
print("q3 is",Q3)

IQR = Q3-Q1
low_lim = Q1 - 1.5*IQR
up_lim = Q3 + 1.5*IQR

print('IQR is',IQR)
print('low_lim is',low_lim)
print('up_lim is',up_lim)
```

```python
#catching the outliers in fare and removing them
outlier = []
for x in df ["matches_played"]:
    if((x>up_lim)or(x<low_lim)):
        outlier.append(x)
```

```python
#spliting into dependant and independant

x = df.drop('selection',axis=1)
y = df['selection']
```

# Creating Training Model

```python
#splitting dataset into train and test   train=20%

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20,random_state=42)
```

# Logistic Regression

```python
from sklearn import linear_model
lr=linear_model.LogisticRegression()
lr_model=lr.fit(x_train,y_train)
```

```python
lr_pred=lr_model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
print(accuracy_score(y_test, lr_pred))
```

logistic regression only got 70% accuracy

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,recall_score,precision_score,classification_report,f1_score
```

```python
metric_k=[]
neighbors=np.arange(3,20)
for k in neighbors:
    classifier=KNeighborsClassifier(n_neighbors=k,metric='euclidean')
    classifier.fit(x_train,y_train)
    y_pred_knn=classifier.predict(x_test)
    acc=accuracy_score(y_test,y_pred_knn)
    metric_k.append(acc)
```

```python
#plotting the graph to find the k value

plt.plot(neighbors,metric_k,'o-')
plt.xlabel('k')
plt.ylabel('accuracy')
plt.grid()
```

the k value 9 has the highest accuracy

```python
knn=KNeighborsClassifier(n_neighbors=9)
knn.fit(x_train,y_train)
y_pred_knn=knn.predict(x_test)
```

```python
confusion_matrix(y_test,y_pred_knn)
```

```python
print('Accuracy= ',accuracy_score(y_test,y_pred_knn))
```

# SVM

```python
from sklearn.svm import SVC
svm_cls=SVC(kernel='rbf')
svm_cls=svm_cls.fit(x_train,y_train)
```

```python
y_pred_svm=svm_cls.predict(x_test)
```

```python
print(accuracy_score(y_test, y_pred_svm))
```
